```python
#19 Prepare X and y
x=df.drop(columns='Species')
y=df['Species']
```

```python
#20 Split into train and test data – 70% train, random state = 1
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
#21 Look at the documentation of the SVC class you’ll use to create the model
from sklearn.svm import SVC
```

```python
#23 Instantiate an object of class SVC using cost C = 1
model = SVC(C=1)
```

```python
#24 Model building
model.fit(x_train,y_train)
```

```python

#25 Predict using the model
y_pred=model.predict(x_test)
```

```python
#26 Evaluate the model using confusion matrix
from sklearn.metrics import confusion_matrix
c_m=confusion_matrix(y_test,y_pred)
print(c_m)
```

```python
#29 k fold cross validation
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, x, y, cv=5)

print("Cross-validation scores:", scores)

print("Average accuracy:", scores.mean())
```

```python
#30 Use grid search CV to find optimal value of C (SVC)
#31 Try with C = [0.1, 1, 10, 100, 1000]
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

grid = GridSearchCV(SVC(), {'C': [0.1, 1, 10, 100, 1000]}, cv=5)
grid.fit(x_train, y_train)

print("Best C:", grid.best_params_['C'])
print("Best Score:", grid.best_score_)
```

```python
#32 Specify number of folds for K-fold CV
model = SVC(C=1)
scores = cross_val_score(model, x, y, cv=10)  # 10-fold CV

print("CV Scores:", scores)
print("Mean Accuracy:", scores.mean())
```
