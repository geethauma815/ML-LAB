```python
import pandas as pd
import numpy as np
df=pd.read_csv("Car_sales.csv")
df["Price in thousands"]=df["Price in thousands"].replace('.',np.nan)
df["Curb weight"]=df["Curb weight"].replace('.',np.nan)
df["Wheelbase"]=df["Wheelbase"].replace('.',np.nan)
df["4-year resale value"]=df["4-year resale value"].replace('.',np.nan)
df["Engine size"]=df["Engine size"].replace('.',np.nan)
df["Horsepower"]=df["Horsepower"].replace('.',np.nan)
df["Width"]=df["Width"].replace('.',np.nan)
df["Length"]=df["Length"].replace('.',np.nan)
df["Fuel capacity"]=df["Fuel capacity"].replace('.',np.nan)
#df["Engine size"]=df["Engine size"].replace('.',np.nan)
df["Fuel efficiency"]=df["Fuel efficiency"].replace('.',np.nan)
#df['Fuel efficiency'].fillna(df['Fuel efficiency'].mean(), inplace=True)


df["Fuel capacity"]=df["Fuel capacity"].astype('float64')
df["Length"]=df["Length"].astype('float64')
df["Width"]=df["Width"].astype('float64')
df["Horsepower"]=df["Horsepower"].astype('float64')
df["Engine size"]=df["Engine size"].astype('float64')
df["4-year resale value"]=df["4-year resale value"].astype('float64')
df["Wheelbase"]=df["Wheelbase"].astype('float64')
df["Curb weight"]=df["Curb weight"].astype('float64')
df["Price in thousands"]=df["Price in thousands"].astype('float64')
df["Fuel efficiency"]=df["Fuel efficiency"].astype('float64')

df['Fuel efficiency'].fillna(df['Fuel efficiency'].mean(), inplace=True)
df['Fuel capacity'].fillna(df['Fuel capacity'].mean(),inplace=True)
df['Price in thousands'].fillna(df['Price in thousands'].mean(),inplace=True)
df['Curb weight'].fillna(df['Curb weight'].mean(),inplace=True)
df['4-year resale value'].fillna(df['4-year resale value'].mean(),inplace=True)
df['Engine size'].fillna(df['Engine size'].mean(),inplace=True)
df['Horsepower'].fillna(df['Horsepower'].mean(),inplace=True)
df['Wheelbase'].fillna(df['Wheelbase'].mean(),inplace=True)
df['Length'].fillna(df['Length'].mean(),inplace=True)
df['Width'].fillna(df['Width'].mean(),inplace=True)
print(df)







```

```python
#14. Plot histogram of continuous numerical variable : price,sales,hoursepower,fuelefficiency
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
# and the columns are named 'Price in thousands', 'Sales in thousands', 'Horsepower', 'Fuel efficiency'

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # 2 rows, 2 columns

# Plot histograms using sns.histplot()
sns.histplot(df['Price in thousands'], bins=10, ax=axes[0, 0])
axes[0, 0].set_title('Price')

sns.histplot(df['Sales in thousands'], bins=10, ax=axes[0, 1])
axes[0, 1].set_title('Sales')

sns.histplot(df['Horsepower'], bins=10, ax=axes[1, 0])
axes[1, 0].set_title('Horsepower')

sns.histplot(df['Fuel efficiency'], bins=10, ax=axes[1, 1])
axes[1, 1].set_title('Fuel Efficiency')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
